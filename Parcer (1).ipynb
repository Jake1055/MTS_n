```python
import scrapy

class FilmsSpider(scrapy.Spider):
    name = 'films'
    start_urls = ['https://ru.wikipedia.org/w/index.php?title=%D0%9A%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D1%8F%3A%D0%A4%D0%B8%D0%BB%D1%8C%D0%BC%D1%8B_%D0%BF%D0%BE_%D0%B0%D0%BB%D1%84%D0%B0%D0%B2%D0%B8%D1%82%D1%83&from=%D0%90/']

    def parse(self, response):
        films_list = response.xpath('//*[@id="mw-pages"]/div/div/div/ul/li')
        for film in films_list:
            film_link = film.xpath('./a/@href').get()
            film_url = 'https://ru.wikipedia.org' + film_link
            yield response.follow(film_url, callback=self.parse_film)
        

    def parse_film(self, response):
       
        producer_l = ['//span[@data-wikidata-property-id="P57"]/a/text()',
                     '//span[@data-wikidata-property-id="P57"]/text()']
        for i in producer_l:
            if response.xpath(i).get() is None:
                continue
            else:
                producer = response.xpath(i).get()
                break
        

        genres_l = ['//span[@data-wikidata-property-id="P136"]/text()',
                    '//span[@data-wikidata-property-id="P136"]/a/text()']
        for i in genres_l:
            if response.xpath(i).get() is None:
                continue
            else:
                genres = response.xpath(i).get()
                break
                
        title = response.xpath('//th[@class="infobox-above"]/text()').get()
        
        countries_links = ['//*[@id="mw-content-text"]/div[1]/table/tbody/tr[16]/td/span/span[1]/a/span/text()',
                           '//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[11]/td/span/span/a/span/text()']
        for i in countries_links:
            if response.xpath(i).get() is None:
                continue
            else:
                country = response.xpath(i).get()
                break
        year_l = ['//span[@class="dtstart"]/text()',
                 '//span[@class="nowrap"]/a/text()']        
        for i in year_l:
            if response.xpath(i).get() is None:
                continue
            else:
                year = response.xpath(i).get()
                break     
        
        yield {
            'Название': title,
            'Жанр': genres,
            'Режисер': producer,
            'Страна': country,
            'Год': year,
        }
from scrapy.crawler import CrawlerProcess
process = CrawlerProcess(settings={'FEEDS': {'output1.csv': {'format': 'csv'}}})
process.crawl(FilmsSpider)
process.start()
```
